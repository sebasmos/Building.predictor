# Tiling imagery and labels using the solaris Python API


**Counting with Imagery and vector labels, create tile vectors at specified fixed-size.**

[Reference] (https://solaris.readthedocs.io/en/latest/tutorials/notebooks/api_tiling_tutorial.html ) We will tile Rio de Janeiro with Solaris


First, downloaded and extracted two files from the spacenet-dataset AWS S3 bucket:

    Imagery: https://s3.amazonaws.com/spacenet-dataset/AOIs/AOI_1_Rio/PS-RGB/PS-RGB_mosaic_013022223133.tif

    Vector labels: https://spacenet-dataset.s3.amazonaws.com/AOIs/AOI_1_Rio/srcData/buildingLabels/Rio_Buildings_Public_AOI_v2.geojson


For this working example we’ll tile into 500-by-500-pixel chips beginning at the top left corner. Note that you can also tile based on the metric units covered by an image - for example, we could specify 250 meter-by-250 meter chips (which is the same size in this case). See the documentation for sol.tile.raster_tile.RasterTiler() for more details.



Initialize the RasterTiler object:


```python
import solaris as sol
import os
```

```python
# Point towards datasets file to avoid overload in github
raster_tiler = sol.tile.raster_tile.RasterTiler(dest_dir='/home/sebasmos/Documentos/TRDP/DATASET/API/API_1/rio_chips',  # the directory to save images to
                                                src_tile_size=(500, 500),  # the size of the output chips
                                                verbose=True)
```

This object can be re-used with the same parameters for multiple images if desired. This way, you can tile multiple images collected over the same geography with the same settings. There are additional arguments that you can provide (for example, the destination coordinate reference system).

To tile the imagery, pass the image file to the tiler’s tile() method, which returns the CRS of the source raster for vector tiling:




```python
raster_bounds_crs = raster_tiler.tile('/home/sebasmos/Documentos/TRDP/DATASET/API/API_1/3band/PS-RGB_mosaic_013022223133.tif')
```

This should throw a few warnings/errors about the input file, which you can ignore. You’ll create 1600 files in your “rio_chips” subdirectory, one for each 500x500 tile - **So what this means is that we have taken the big image tile and splited into smaller chunks of imagery of 500x500 size with CRS for vector tiling**. The filenames are in the format [src-filename]\_[longitude]\_[latitude].tif. Reprojection takes a while, so be patient.

Once that process finishes, we’ll use these auto-generated tile boundaries, which are stored in raster_tiler, to create vector tiles.

These bounds are in the format [left, bottom, right, top] in the input file CRS. The following line prints the first set of bounds (there are 1600 in the list):

```python
print(raster_tiler.tile_bounds[0])
```

`raster_tiler.tile_bounds` is passed as an argument into the `VectorTiler instance.`

```python
vector_tiler = sol.tile.vector_tile.VectorTiler(dest_dir='/home/sebasmos/Documentos/TRDP/DATASET/API/API_1/rio_labels',
                                                verbose=True)
vector_tiler.tile('/home/sebasmos/Documentos/TRDP/DATASET/API/API_1/geojson/Rio_Buildings_Public_AOI_v2.geojson',
                  tile_bounds=raster_tiler.tile_bounds,
                  tile_bounds_crs=raster_bounds_crs)
```

**Remarks**

* At the end we count with 4 folders: 3band(containing .tile imagery), geojson(containing JSON file), rio chips (chunks of vectorized tiles) and rio labels (which it's is just the same geojson file but splitted for all the chips, correspondingly for each sub-tile). 

* 3band/ contains `PS-RGB_mosaic_013022223133.tiff` - high resolution image

* geojson/ Contains `JSON` file with satellite main features stored in dictionary and CRS (type, name, properties and ` "name": "urn:ogc:def:crs:OGC:1.3:CRS84"`). Properties include: timestamp, version, changeset, user, uid, HGIS_OID, building, type (such as polygon), id, relation, area, QAStatus, HGISOID, TaskArea, revision1, , Shape_Leng, Shape_Area, geometry, and coordinates. The same for every single tile.

* Riochips/ contains invididual 

Therefore, imagery is splited into training and target sets, based on CRS, fixed size as desired and fixed metadata for each sub-tile.
